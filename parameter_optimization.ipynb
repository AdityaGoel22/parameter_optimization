```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import random as r
from sklearn.svm import SVC
from sklearn import svm
from sklearn.svm import NuSVC
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/kaggle/input/dry-beans/Dry Beans Dataset.csv")
```

```python
data.head()
```

```python
data.isna().sum()
```

```python
accuracy_list = []
iteration_list = []
bestAccuracy = 0
bestKernel = ""
bestNu = 0
bestEpsilon = 0
iteration = 100
kernelList = ['poly', 'rbf', 'precomputed', 'linear', 'sigmoid']
```

```python
X = data.drop(columns = ["Class"], axis = 1)
y = data['Class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=42)
```

```python
for i in range (iteration):
    k = r.sample(kernelList,1),
    n = r.random(),
    e = r.random(),

    model = svm(kernel = k, nu = n, epsilon = e)
    model.fit(X_train, y_train)
    pred = round(model.predict(X_test))
    accuracy = accuracy_score(y_test,pred)
    accuracy = round(accuracy)
    
    accuracy_list.append(accuracy)
    iteration_list.append(i)


    if accuracy > bestAccuracy:
        bestKernel = k
        bestNu = n
        bestEpsilon = e
        bestAccuracy = accuracy
```

```python
plt.plot(iteration_list, accuracy_list)
plt.xlabel('Iterations')
plt.ylabel('Accuracy')
plt.title('Fiteness (best Accuracy)')
plt.show()
```
